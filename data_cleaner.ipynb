```python
import pandas as pd

def clean_client_data():
    print("Loading broken client file...")
    # Loading file
    df = pd.read_csv('messy_client_data.csv')
    
    # Standardizing Column Names
    df.columns = df.columns.str.strip().str.upper()
    
    # Handling Critical Missing Values
    df = df.dropna(subset=['TRANSACTION_ID'])
    
    # Cleaning IDs and Removing Duplicates
    df['TRANSACTION_ID'] = df['TRANSACTION_ID'].astype(str).str.strip().str.upper()
    df = df.drop_duplicates(subset=['TRANSACTION_ID'], keep='last')
    
    # Cleaning Text Data
    df['CUSTOMER_NAME'] = df['CUSTOMER_NAME'].fillna('Unknown')
    df['CUSTOMER_NAME'] = df['CUSTOMER_NAME'].astype(str).str.strip().str.title()
    
    # Cleaning Numeric Data 
    df['ITEM_PRICE'] = df['ITEM_PRICE'].astype(str).str.replace('$', '', regex=False).str.strip()
    df['ITEM_PRICE'] = pd.to_numeric(df['ITEM_PRICE'], errors='coerce') 
    df['ITEM_PRICE'] = df['ITEM_PRICE'].fillna(0.0) 
    
    # Cleaning Dates
    df['PURCHASE_DATE'] = pd.to_datetime(df['PURCHASE_DATE'], errors='coerce', format='mixed')
    
    # Fixing dates
    df['PURCHASE_DATE'] = df['PURCHASE_DATE'].ffill()
    df['PURCHASE_DATE'] = df['PURCHASE_DATE'].dt.strftime('%Y-%m-%d')
    
    # Saving
    df.to_csv('clean_client_data.csv', index=False)
    print("✅ Pipeline complete! Flawless data saved as 'clean_client_data.csv'")
    
    return df


clean_df = clean_client_data()
display(clean_df)
```
